```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("product_sales.csv")
```

## Data Validation and Cleaning
Krótka sekcja porządkująca dane przed analizą – upewniamy się, że typy są poprawne, a zbiór jest wolny od duplikatów i braków.

```python
df.describe(include='all', datetime_is_numeric=True).T
```

```python
# Standaryzacja typów danych
numeric_cols = ['week', 'nb_sold', 'years_as_customer', 'nb_site_visits']
df[numeric_cols] = df[numeric_cols].astype('int64')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['customer_id'] = df['customer_id'].astype('string')
df['week_start'] = pd.to_datetime('2023-01-02') + pd.to_timedelta(df['week'] - 1, unit='W')
```

```python
# Usunięcie duplikatów i braków danych
before_rows = len(df)
df = df.drop_duplicates()
after_rows = len(df)
duplicates_removed = before_rows - after_rows
missing_before = df['revenue'].isna().sum()
df = df.dropna(subset=['revenue'])
missing_after = df['revenue'].isna().sum()
print(f'Duplicates removed: {duplicates_removed}')
print(f'Missing revenue before: {missing_before}, after cleaning: {missing_after}')
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['revenue'], color='skyblue')
plt.title('Boxplot przychodu (Revenue) po czyszczeniu danych')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# 3) KPI per metoda
summary = df.groupby('sales_method').agg(
    customers=('customer_id','nunique'),
    conversion=('revenue', lambda s: (s>0).mean()),
    arpc=('revenue','mean')
).reset_index()
print(summary)
```

```python
# 4) Czas pracy per metoda (godziny/klient)
time_map = {'Emails': 0.0, 'Calls': 0.5, 'Email and Call': 1/6}  # 0.1667 h
df['hours_per_customer'] = df['sales_method'].map(time_map)
```

```python
eff = df.groupby('sales_method').agg(
    total_revenue=('revenue','sum'),
    total_hours=('hours_per_customer','sum')
)
eff['RPSH'] = eff['total_revenue'] / eff['total_hours'].replace(0, pd.NA)
kpi = summary.merge(eff.reset_index(), on='sales_method', how='left')
```

## 1/ Sprzedarz tygodniowa
---

```python
sns.histplot(df['week'], bins=6)
plt.title('Liczba sprzedaży w poszczególnych tygodniach')
plt.xlabel('Tydzień od premiery')
plt.ylabel('Liczba transakcji (klientów)')
plt.tight_layout()
plt.show()
```

---
**Wnioski (PL):** Większość transakcji przypada na pierwszy tydzień po premierze (~3,7 tys. klientów), a następnie wolumen maleje do około 1,2 tys. w tygodniu 6 z jedynie krótkim odbiciem w tygodniach 4–5, co wskazuje na szybko wygasające zainteresowanie.

**Findings (EN):** Most purchases arrive in launch week (~3.7k customers) before volumes slide to ~1.2k by week 6 with only a brief rebound in weeks 4–5, signalling that post-launch demand decays quickly.


```python
sns.boxplot(data=df, x='week', y='revenue')
plt.title('Rozkład przychodów (Revenue)')
plt.xlabel('week')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

---
**Wnioski (PL):** Rozkład przychodów przesuwa się w górę z każdym tygodniem — mediana i górne kwartyle rosną, a w tygodniu 6 pojawiają się najwyższe wartości i większa zmienność, co sugeruje coraz wyższe wydatki na klienta.

**Findings (EN):** The revenue distribution shifts upward week by week — medians and upper quartiles rise and week 6 shows both the highest levels and wider spread, implying growing spend per customer over time.


```python
from sklearn.linear_model import LinearRegression

trend = df.groupby(['week', 'sales_method'])['revenue'].mean().reset_index()

preds = []

for method in trend['sales_method'].unique():
    subset = trend[trend['sales_method'] == method]
    
    X = subset['week'].values.reshape(-1, 1)
    y = subset['revenue'].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    # predykcja dla tygodni istniejących + tygodnia 7
    future_weeks = np.arange(1, 8).reshape(-1, 1)
    y_pred = model.predict(future_weeks)
    
    # zapisz do listy
    temp = pd.DataFrame({
        'week': future_weeks.flatten(),
        'predicted_revenue': y_pred,
        'sales_method': method
    })
    preds.append(temp)

# scal wyniki
pred_df = pd.concat(preds, ignore_index=True)
```

```python
sns.lineplot(data=trend, x='week', y='revenue', hue='sales_method', marker='o', linestyle='--', alpha=0.6)
sns.lineplot(data=pred_df, x='week', y='predicted_revenue', hue='sales_method', marker='o')

plt.title('Predykcja średniego przychodu w czasie (week × sales_method)')
plt.xlabel('Tydzień od premiery produktu')
plt.ylabel('Średni revenue (prognoza)')
plt.legend(title='Metoda sprzedaży')
plt.tight_layout()
plt.show()
```

---
**Wnioski (PL):** Średni przychód rośnie we wszystkich kanałach, najsilniej w strategii Email and Call, a prosta prognoza wydłużona do tygodnia 7 utrzymuje trend wzrostowy, co przemawia za kontynuacją kampanii wielokanałowej.

**Findings (EN):** Average revenue climbs across every channel, most steeply for the Email and Call mix, and the week-7 projection keeps the upward slope, reinforcing the value of sustaining the multichannel push.
